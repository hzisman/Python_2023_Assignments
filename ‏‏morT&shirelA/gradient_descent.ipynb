#### <font color='lightblue'>Cost function <font/>
##### Defining a cost function to calculate the mean squared error
    

\begin{equation*}
\left (\sum_{i=1}^n (mx_i-y_i)^2 \right)/n
\end{equation*}
    

```python
def cost(m):
    new=((x*m)-y)**2
    cnt=len(x)
    return new.sum()/cnt
```

```python
cost(3)
```

#### <font color='lightblue'>d_cost <font/>
#####  Defining the derivative of the cost function with respect to 'm'

\begin{equation*}
\left (\sum_{i=1}^n 2X_i(mx_i-y_i) \right)/n
\end{equation*}

```python
def dcost(m):
    dcst= 2*(m*x-y)*x
    return dcst.sum()/len(x) 
```

```python
import random

history = []
step = 0.0001
for i in range(1000):
    print(m)
    m_dcost = dcost(m)
    history.append(cost(m))
    m -= step * (m_dcost)
```
